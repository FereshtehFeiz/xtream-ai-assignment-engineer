```python
#Dropping dimentionless diamonds
data = data.drop(data[data["x"]==0].index)
data = data.drop(data[data["y"]==0].index)
data = data.drop(data[data["z"]==0].index)
data.shape
```

```python
data.shape
```

```python
# Scatter plot of carat vs. price
data.plot(kind='scatter', x='carat', y='price', alpha=0.5, color='blue')
plt.title('Carat vs. Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()

# Box plot of cut vs. price
data.boxplot(column='price', by='cut')
plt.title('Boxplot of Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()

# Box plot of color vs. price
data.boxplot(column='price', by='color')
plt.title('Boxplot of Price by Color')
plt.xlabel('Color')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter plot of x vs. price
data.plot(kind='scatter', x='x', y='price', alpha=0.5, color='green')
plt.title('x vs. Price')
plt.xlabel('x')
plt.ylabel('Price')
plt.show()

# Scatter plot of y vs. price
data.plot(kind='scatter', x='y', y='price', alpha=0.5, color='red')
plt.title('y vs. Price')
plt.xlabel('y')
plt.ylabel('Price')
plt.show()

# Scatter plot of z vs. price
data.plot(kind='scatter', x='z', y='price', alpha=0.5, color='orange')
plt.title('z vs. Price')
plt.xlabel('z')
plt.ylabel('Price')
plt.show()

# Scatter plot of depth vs. price
data.plot(kind='scatter', x='depth', y='price', alpha=0.5, color='purple')
plt.title('Depth vs. Price')
plt.xlabel('Depth')
plt.ylabel('Price')
plt.show()

# Scatter plot of table vs. price
data.plot(kind='scatter', x='table', y='price', alpha=0.5, color='blue')
plt.title('Table vs. Price')
plt.xlabel('Table')
plt.ylabel('Price')
plt.show()
```

```python
ax=sns.pairplot(data, hue= "cut")
```

```python
#Dropping the outliers. 
data = data[(data["depth"]<75)&(data["depth"]>45)]
data = data[(data["table"]<80)&(data["table"]>40)]
data = data[(data["x"]<30)]
data = data[(data["y"]<30)]
data = data[(data["z"]<30)&(data["z"]>2)]
data.shape
```

```python
# Scatter plot of x vs. price
data.plot(kind='scatter', x='x', y='price', alpha=0.5, color='green')
plt.title('x vs. Price')
plt.xlabel('x')
plt.ylabel('Price')
plt.show()

# Scatter plot of y vs. price
data.plot(kind='scatter', x='y', y='price', alpha=0.5, color='red')
plt.title('y vs. Price')
plt.xlabel('y')
plt.ylabel('Price')
plt.show()

# Scatter plot of z vs. price
data.plot(kind='scatter', x='z', y='price', alpha=0.5, color='orange')
plt.title('z vs. Price')
plt.xlabel('z')
plt.ylabel('Price')
plt.show()

# Scatter plot of depth vs. price
data.plot(kind='scatter', x='depth', y='price', alpha=0.5, color='purple')
plt.title('Depth vs. Price')
plt.xlabel('Depth')
plt.ylabel('Price')
plt.show()

# Scatter plot of table vs. price
data.plot(kind='scatter', x='table', y='price', alpha=0.5, color='blue')
plt.title('Table vs. Price')
plt.xlabel('Table')
plt.ylabel('Price')
plt.show()
```

```python
# Get list of categorical variables
s = (data.dtypes =="object")
object_cols = list(s[s].index)
print("Categorical variables:")
print(object_cols)
```

```python

# Violin plot of cut vs. price
sns.violinplot(data=data, x='cut', y='price')
plt.title('Violin plot of Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()

# Violin plot of color vs. price
sns.violinplot(data=data, x='color', y='price')
plt.title('Violin plot of Price by Color')
plt.xlabel('Color')
plt.ylabel('Price')
plt.show()

# Violin plot of clarity vs. price
sns.violinplot(data=data, x='clarity', y='price')
plt.title('Violin plot of Price by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Price')
plt.show()

#These violin plots display the distribution of prices within each category of the categorical variables (cut, color, clarity).
#The width of the violin plot represents the density of prices at different values, and the white dot inside each violin represents the median price.
```

```python
from sklearn.preprocessing import LabelEncoder
# Make copy to avoid changing original data 
label_data = data.copy()

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols:
    label_data[col] = label_encoder.fit_transform(label_data[col])
label_data.head()
```

```python
#correlation matrix
cmap = sns.diverging_palette(70,20,s=50, l=40, n=6,as_cmap=True)
corrmat= label_data.corr()
f, ax = plt.subplots(figsize=(12,12))
sns.heatmap(corrmat,cmap=cmap,annot=True, )
```

Points to notice:

"x", "y" and "z" show a high correlation to the target column.
